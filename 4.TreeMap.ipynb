```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import DBSCAN
import plotly.express as px
import datetime
```

```python
sample_df = pd.read_csv('NLP_Dataset1.csv')
```

```python
sample_df.head()
```

```python
grouped_df = sample_df.groupby(['organizationcountrycode', 'Severity','AssetType']).size().reset_index(name='count')

# Display the grouped DataFrame
print(grouped_df.head())
```

```python
grouped_df.dtypes
```

```python
# Ensure correct data types
grouped_df['organizationcountrycode'] = grouped_df['organizationcountrycode'].astype(str)
grouped_df['Severity'] = grouped_df['Severity'].astype(str)
grouped_df['AssetType'] = grouped_df['AssetType'].astype(str)
grouped_df['count'] = grouped_df['count'].astype(int)
```

```python
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd


# Create the Dash app
app = JupyterDash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1("Treemap of Alarms Clustered by Country, Severity, and Asset Type"),
    
    html.Label("Select Country:"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in grouped_df['organizationcountrycode'].unique()],
        multi=True,
        value=list(grouped_df['organizationcountrycode'].unique())
    ),
    
    html.Label("Select Severity:"),
    dcc.Dropdown(
        id='severity-dropdown',
        options=[{'label': severity, 'value': severity} for severity in grouped_df['Severity'].unique()],
        multi=True,
        value=list(grouped_df['Severity'].unique())
    ),
    
    dcc.Graph(id='treemap-plot')
])

# Callback to update the treemap plot based on selected filters
@app.callback(
    Output('treemap-plot', 'figure'),
    [Input('country-dropdown', 'value'),
     Input('severity-dropdown', 'value')]
)
def update_treemap(selected_countries, selected_severities):
    filtered_df = grouped_df[(grouped_df['organizationcountrycode'].isin(selected_countries)) & 
                             (grouped_df['Severity'].isin(selected_severities))]
    
    fig = px.treemap(
        filtered_df,
        path=[px.Constant('World'),'organizationcountrycode', 'Severity', 'AssetType'],
        values='count',  # This will count the number of rows per category
        title='Treemap of Alarms Clustered by Country, Severity, and Asset Type',
        labels =['count'],
        hover_data=['count']  # Include the count in hover data
    )

    fig.update_layout(
        treemapcolorway=['green','#FF0000', '#FF8C00', '#FFD700', '#ADFF2F', '#00FF00']
    )

    return fig
import socket

def find_free_port():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.bind(('', 0))  # Bind to a free port provided by the host
    port = s.getsockname()[1]
    s.close()
    return port

# Find an available port
available_port = find_free_port()
#print(f"Available port: {available_port}")
# Run the app
app.run_server(mode='inline', port=find_free_port())
```
